# Color Space Conversion
- ### RGB Separated channels 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('person.jpg')
```

```python
def get_channel(value):

    if value == 0:
        b = src.copy()
        # set green and red channels to 0
        b[:, :, 1] = 0
        b[:, :, 2] = 0
        return b
    
    elif value == 1:
        g = src.copy()
        # set blue and red channels to 0
        g[:, :, 0] = 0
        g[:, :, 2] = 0
        return g
    
    else:
        r = src.copy()
        # set blue and green channels to 0
        r[:, :, 0] = 0
        r[:, :, 1] = 0
        return r

```

```python
blue_img = get_channel(0)
green_img = get_channel(1)
red_img = get_channel(2)


plt.figure(figsize = (14,4))
plt.subplot(1, 4, 1)
plt.imshow(blue_img)
plt.subplot(1, 4, 2)
plt.imshow(green_img)
plt.subplot(1, 4, 3)
plt.imshow(red_img)
plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
```

```python
plt.figure(figsize = (14,4))

plt.subplot(1, 4, 1)
plt.title("blue_img - green_img")
plt.imshow(blue_img - green_img)

plt.subplot(1, 4, 2)
plt.title("blue_img - red_img")
plt.imshow(blue_img - red_img)

plt.subplot(1, 4, 3)
plt.title("green_img - red_img")
plt.imshow(green_img - red_img)

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
```

```python
plt.figure(figsize = (14,4))

plt.subplot(1, 4, 1)
plt.title("green_img - blue_img")
plt.imshow(green_img - blue_img)

plt.subplot(1, 4, 2)
plt.title("red_img - blue_img")
plt.imshow(red_img - blue_img)

plt.subplot(1, 4, 3)
plt.title("red_img - green_img")
plt.imshow(red_img - green_img)

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
```

- ### RGB to HSV and YUV

```python
yuv = cv2.cvtColor(src,cv2.COLOR_RGB2YUV)
hsv = cv2.cvtColor(src,cv2.COLOR_RGB2HSV)

plt.figure(figsize = (12,4))

plt.subplot(1, 3, 1)
plt.title("YUV")
plt.imshow(yuv)
plt.subplot(1, 3, 2)
plt.title("HSV")
plt.imshow(hsv)
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
```
